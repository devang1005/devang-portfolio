## Q8.  Which product recorded the highest Incremental Revenue Percentage (IR%) during the Sankranti campaign? What is the IR% for this produc

```python
df_sankranti = df_fact_events[df_fact_events['campaign_id'] == 'CAMP_SAN_01']
df_sankranti.shape
```

```python
df_sankranti.duplicated().sum()
```

```python
df_sankranti = df_sankranti.drop_duplicates()
df_sankranti.shape
```

```python
df_sankranti.isnull().sum()
```

```python
df_fact_events['quantity_sold(before_promo)'].median()
```

```python
df_sankranti['quantity_sold(before_promo)'] = df_sankranti['quantity_sold(before_promo)'].fillna(78)
```

```python
df_sankranti.isnull().sum()
```

```python
df_sankranti.shape
```

```python
df_sankranti.groupby('product_code')['base_price(before_promo)'].unique().reset_index()
```

```python
df_grouped_After_promo_product_id = df_sankranti.groupby('product_code')['quantity_sold(after_promo)'].sum().reset_index()
df_grouped_After_promo_product_id
```

```python
# df_sankranti.groupby('product_code')['base_price(after_promo)'].unique().reset_index()
```

```python
df_price_after_promo = df_sankranti.groupby('product_code')['base_price(after_promo)'].first().reset_index()
df_price_after_promo
```

```python
df_grouped_After_promo_product_id['base_price(after_promo)'] = df_price_after_promo['base_price(after_promo)']
df_grouped_After_promo_product_id.head()
df_grouped_After_promo_product_id['revenue_after_promo'] = df_grouped_After_promo_product_id['quantity_sold(after_promo)']*df_grouped_After_promo_product_id['base_price(after_promo)']
df_grouped_After_promo_product_id.head()
```

```python
df_grouped_product_id = df_sankranti.groupby('product_code')['quantity_sold(before_promo)'].sum().reset_index()
df_grouped_product_id
```

```python
df_base_price = df_sankranti.groupby('product_code')['base_price(before_promo)'].first().reset_index()
df_grouped_product_id['product_base_price_before_promo'] = df_base_price['base_price(before_promo)']
df_grouped_product_id.head()
df_grouped_product_id['revenue_before_promo'] = df_grouped_product_id['quantity_sold(before_promo)']*df_grouped_product_id['product_base_price_before_promo']
df_grouped_product_id.head()
df_grouped_product_id['revenue_after_promo'] = df_grouped_After_promo_product_id['revenue_after_promo']
df_grouped_product_id.head()
```

```python
df_grouped_product_id['revenue_before_promo'] = df_grouped_product_id['quantity_sold(before_promo)']*df_grouped_product_id['product_base_price_before_promo']
df_grouped_product_id.head()
```

```python
df_grouped_product_id.head()
```

```python
df_grouped_product_id['IR%'] = (df_grouped_product_id['revenue_after_promo'] - df_grouped_product_id['revenue_before_promo'])/df_grouped_product_id['revenue_before_promo']
df_grouped_product_id
```

```python
highest_IR_val = df_grouped_product_id['IR%'].max()
print(highest_IR_val*100)
```

```python
df_grouped_product_id[df_grouped_product_id['IR%'] == highest_IR_val]
```
